```python
import os

import pandas as pd

import csv

import shutil
```

```python
'''
TESTING THE SCRIPT
'''
```

```python
# Test the script on ONE participant first
base_directory = '/scratch/a/arisvoin/clarasun/SCanD_project_PNC/data/local/old_ciftify'
```

```python
# Load the list of participants from the TSV file
participants = [] # try list instead of set()
tsv_file = 'subjects_with_qc_fmri.tsv'

with open(tsv_file, 'r') as tsvfile:
    reader = csv.reader(tsvfile, delimiter='\t')
    for row in reader:
        # Assuming the participant names are in the first column of the TSV file
        participant_name = row[0]
        participants.append(participant_name)
```

```python
participants = participants[1:]
# remove the 'participant_id' header row from the list
```

```python
test_participant = participants[:1]
```

```python
test_participant
# list with only sub-600515650925
```

```python
# Iterate through the base directory and delete participant folders that are one level below
for entry in os.listdir(base_directory):
    if entry in test_participant:
        entry_path = os.path.join(base_directory, entry)
        try:
            print(f"Deleting folder: {entry_path}")
            # Use shutil.rmtree() for recursive folder deletion
            # Make sure to test with a small set of folders before uncommenting the next line
            shutil.rmtree(entry_path)
        except Exception as e:
            print(f"Error deleting {entry_path}: {e}")
```

```python
# Define the base directory
base_directory = '/scratch/a/arisvoin/clarasun/SCanD_project_PNC/data/local/old_ciftify'

# Iterate through the base directory and delete participant folders that are in the list
for entry in os.listdir(base_directory):
    if entry in participants:
        entry_path = os.path.join(base_directory, entry)
        try:
            print(f"Deleting folder: {entry_path}")
            # Use shutil.rmtree() for recursive folder deletion
            # Make sure to test with a small set of folders before uncommenting the next line
            shutil.rmtree(entry_path)
        except Exception as e:
            print(f"Error deleting {entry_path}: {e}")
```
