```python
'''
STEP ONE
'''
# remove first 40 participants (rows) from participants_original.tsv

# load tsv
tsv_og = pd.read_csv('/scratch/a/arisvoin/clarasun/SCanD_project_PNC/data/local/bids/participants_original.tsv', sep='\t')

# remove the first 40 rows
tsv1 = tsv_og.iloc[40:] # use iloc slicing method to remove rows
# remember that indexing starts at 0 in python 
tsv1
```

```python
'''
STEP TWO
'''
# merge tsv1 from step 1 with new_subjects_qc_fmri.tsv 
# exclude all participants in the new_subjects_qc_fmri.tsv

# load tsv
tsv2 = pd.read_csv('/scratch/a/arisvoin/clarasun/SCanD_project_PNC/data/local/bids/new_subjects_qc_fmri.tsv', sep='\t')

# filter tsv1 to exclude all rows of tsv1 that are in tsv2
filtered_tsv1 = tsv1[~tsv1['participant_id'].isin(tsv2['participant_id'])]
filtered_tsv1
# expect there to be 744 rows in filtered_tsv1 because:
# started with 1598 participants
# subtract 40 that were accidentally deleted = 1598 - 40 = 1558
# subtract 814 that have qc_fmri outputs = 1558 - 814 = 744
```

```python
'''
STEP THREE
'''
# filter the resulting tsv from step 2 with missing_bids_func.tsv 
# exclude all participants in missing_bids_func.tsv

# load tsv
tsv3 = pd.read_csv('/scratch/a/arisvoin/clarasun/SCanD_project_PNC/data/local/bids/missing_bids_func.tsv', sep='\t')

# filter the filtered_tsv1 to exclude all rows that are in tsv3
new_filtered_tsv1 = filtered_tsv1[~filtered_tsv1['participant_id'].isin(tsv3['participant_id'])]
new_filtered_tsv1
# expect there to be ~544 rows in filtered_tsv1 because:
# started with 744 from filtered_tsv1
# subtract 200 are missing bids func folders
```
